<h3 style=color:blue > Preprocesado de datos para su uso </h3> 

#### Primero aplicamos un test de shapiro a todas las columnas para saber si siguen una distribución gaussiana.

```python
from scipy.stats import shapiro

for i in df:
    stat, p = shapiro(df[i])
    print('nombre de la variable', i)
    print('stat=%.3f, p=%.3f' % (stat, p))
    
    if p > 0.05:
        print('distribucion gaussiana', "\n")
    else:
        print('distribucion NO gaussiana', "\n")
```

#### Todas las variables siguen una distribucion no gaussiana excepto la variable Alcalinity of Ash por lo que esta debera ser estandarizada con otro metodo

#### Separamos las columas segun si las variables cuentan con outliers, si no cuentan con outliers ademas del caso de la variable gaussiana

```python
outliers_variables = ['Malic Acid', 'Ash','Proanthocyanins','Color Intensity','Hue'] 
no_outliers_variables = ['Alcohol', 'Total Phenols', 'Flavanoids', 'Non Flavanoid Phenols', 'od280/od315_of_diluted_wines']
variable_gaussiana = ['Alcalinity of Ash']
```

#### Aplicamos RobustScaler a las variables con outliers

```python
from sklearn.preprocessing import RobustScaler

scaling_robust = RobustScaler()

df_robust = scaling_robust.fit_transform(df[outliers_variables])

df_robust_scaler = pd.DataFrame(df_robust, columns=outliers_variables)

df_robust_scaler.head(5)
```

#### Aplicamos MinMaxScaler a las variables que no tienen outliers

```python
from sklearn.preprocessing import MinMaxScaler


scaling_MinMax = MinMaxScaler()

df_MinMax = scaling_MinMax.fit_transform(df[no_outliers_variables])

df_MinMax = pd.DataFrame(df_MinMax, columns=no_outliers_variables)

df_MinMax
```

#### Por ultimo aplicamos StandardScaler a la variable que sigue una distribucion gaussiana.

```python
from sklearn.preprocessing import StandardScaler

scaling_StandardScaler = StandardScaler()

df_StandardScaler = scaling_StandardScaler.fit_transform(df[variable_gaussiana])

df_StandardScaler = pd.DataFrame(df_StandardScaler, columns=variable_gaussiana)

df_StandardScaler
```

```python
df_final = pd.concat([df['Class'], df_MinMax.reset_index(drop=True), df_StandardScaler.reset_index(drop=True), df_robust_scaler.reset_index(drop=True)], axis=1)
df_final
```

#### Concatenamos todos nuestros DFs en un ultimo para aplicar la separacion en Train y Test

```python
fig = plt.figure(figsize=(20,14))
sns.boxplot(df_final, orient='h')
plt.title('df_final')
```

#### Luego de una nueva comprobacion mediante BoxPlot vemos que los datos han sido normalizados y estandarizados de forma correcta ya que las variabales con outliers se encuentran distribuidas cercanas al 0, mientras que las variables sin outliers se encuentran distribuidas mayoritariamente entre 0 y 1.

#### Support es el número real de instancias en cada clase en el conjunto de prueba para nuestras clases es de 14, 14 y 8 respectivamente.

#### Macro avg y Weighted Avg Son promedios de las métricas para todas las clases.

```python
accuracy_svc = accuracy_score(y_test, y_svc_predict)
print(f'Accuracy SVC: {accuracy_svc}', "\n")
print('Classification Report SVC:', "\n")
print(classification_report(y_test, y_svc_predict))
```

## Comparación mediante Learning-Curve

```python
from sklearn.model_selection import learning_curve

# Curva de Aprendizaje para KNN

train_sizes = np.linspace(0.1, 1.0, 10)
train_sizes_abs, train_scores, test_scores = learning_curve(knn, X_train, y_train, train_sizes=train_sizes, cv=5)
train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

# Curva de Aprendizaje para Support Vector Machine 
train_sizes_abs_svc, train_scores_svc, test_scores_svc = learning_curve(svc, X_train, y_train, train_sizes=train_sizes, cv=5)
train_mean_svc = np.mean(train_scores_svc, axis=1)
test_mean_svc = np.mean(test_scores_svc, axis=1)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Primera subfigura: Curva de Aprendizaje para KNN
axes[0].plot(train_sizes_abs, train_mean, label='Precisión en entrenamiento')
axes[0].plot(train_sizes_abs, test_mean, label='Precisión en prueba')
axes[0].set_xlabel('Tamaño del conjunto de entrenamiento')
axes[0].set_ylabel('Precisión')
axes[0].set_title('Curva de Aprendizaje (K Neighbors)')
axes[0].legend()

# Segunda subfigura: Curva de Aprendizaje para Support Vector Machine (SVM)
axes[1].plot(train_sizes_abs_svc, train_mean_svc, label='Precisión en entrenamiento')
axes[1].plot(train_sizes_abs_svc, test_mean_svc, label='Precisión en prueba')
axes[1].set_xlabel('Tamaño del conjunto de entrenamiento')
axes[1].set_ylabel('Precisión')
axes[1].set_title('Curva de Aprendizaje (Support Vector Machine)')
axes[1].legend()

plt.tight_layout() 
plt.show()
```

#### Ambos modelos mostraron una tasa de clasificacion alta, al igual que han mostrado trabajar mejor a medida que el conjunto de datos aumenta. Por otra parte en cuanto al modelo SVM podemos ver que la puntuacion en el entrenamiento es alta aunque el conjunto de datos sea pequeño y tiende a estabilizarse a medida que el conjunto de datos aumenta, esto puedo significar que no hace falta agregar nuevos datos al modelo, en el caso de la precision en el conjunto de prueba, comienza con menos de un 60% y comienza a trabajar mucho mejor a medida que el conjunto de datos aumenta.

### Entrena’ls usant els diferents paràmetres que admeten per tal de millorar-ne la predicció.

```python
from sklearn.model_selection import GridSearchCV
```

### Obtenemos los parametros de KNN

```python
knn.get_params()
```

```python
from sklearn.model_selection import GridSearchCV, KFold

# Definir el espacio de búsqueda
parametros_ajustados_knn = {
    'n_neighbors': [3, 5, 7, 9],  # Nro de vecinos
    'weights': ['uniform', 'distance'],  # Peso de los vecinos
    'p': [1, 2],  # Variacion del parametro P
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],  # Diferentes tipos de algoritmos 
    'leaf_size': [10, 20, 30, 40],  # Tamaño de la hoja
    'metric': ['euclidean', 'manhattan', 'minkowski'],  # Métricas de distancia
}

# Configurar la búsqueda en cuadrícula con validación cruzada
grid_search_knn = GridSearchCV(knn, parametros_ajustados_knn, cv=KFold(n_splits=5, shuffle=True, random_state=42),scoring='accuracy', n_jobs=-1)

# Realizar la búsqueda en cuadrícula
grid_search_knn = grid_search_knn.fit(X_train, y_train)


print("Mejores parámetros:", grid_search_knn.best_params_)
print("Mejor rendimiento (accuracy):", grid_search_knn.best_score_)
```

#### Luego de obtener los mejores parametros para Knn y obtener un accuracy de 97,8% con los estos pasamos a asignar los nuevos parametros a nuestro modelo antiguo.

```python
knn_new_param = KNeighborsClassifier(algorithm='auto', leaf_size=10, metric='manhattan', n_neighbors = 3, p = 1, weights = 'uniform')
knn_new_predict = knn_new_param.fit(X_train, y_train)
```

```python
knn_new_predict = knn_new_predict.predict(X_test)
knn_new_predict
```

```python
knn_new_accuracy = accuracy_score(y_test, knn_new_predict)
print(f'Accuracy KNN: {knn_new_accuracy}', "\n")
print('Classification Report KNN:', "\n")
print(classification_report(y_test, knn_new_predict))
```

#### Luego de entrenar nuestro modelo con los nuevos parametros podemos ver que el modelo paso de tener un accuracy de 86% a tener un 91,6% por lo que encontramos una mejoria en la precicion del modelo luego de cambiar los parametros.

### obtenemos los parametros para SVC

```python
svc.get_params()
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Definir el espacio de búsqueda
parametros_ajustados_svc = {
    'C': [0.1, 1, 10, 100],  # Variar el parámetro de regularización C
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # Probar diferentes tipos de kernel
    'degree': [2, 3, 4],  # Solo si estás utilizando kernel polinómico
    'gamma': ['scale', 'auto', 0.1, 1],  # Variar el coeficiente del kernel ('scale' y 'auto' son opciones automáticas)
    'class_weight': [None, 'balanced'],  # Puedes variar el peso de las clases
    'shrinking': [True, False],  # Probar con o sin encogimiento
    'probability': [True, False],  # Evaluar la probabilidad de las clases
}

# Configurar la búsqueda en cuadrícula con validación cruzada
grid_search_svc = GridSearchCV(svc, parametros_ajustados_svc, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42))

# Realizar la búsqueda en cuadrícula
grid_search_svc = grid_search_svc.fit(X_train, y_train)


print("Mejores parámetros:", grid_search_svc.best_params_)
print("Mejor rendimiento (accuracy):", grid_search_svc.best_score_)
```

#### Luego de obtener los mejores parametros para SVC y obtener un accuracy de 96,5% con los estos pasamos a asignar los nuevos parametros a nuestro modelo antiguo.

```python
svc_new_param = SVC(C=10, class_weight = None, degree = 2, gamma = 'scale', kernel = 'rbf', probability = True, shrinking = True)
svc_new_predict = svc_new_param.fit(X_train, y_train)
```

```python
svc_new_predict = svc_new_predict.predict(X_test)
svc_new_predict
```

```python
svc_new_accuracy = accuracy_score(y_test, svc_new_predict)
print(f'Accuracy KNN: {svc_new_accuracy}', "\n")
print('Classification Report KNN:', "\n")
print(classification_report(y_test, svc_new_predict))
```

#### Luego de entrenar nuestro modelo con los nuevos parametros podemos ver que el modelo paso de tener un accuracy de 91,6% a tener un 94,4% por lo que encontramos una mejoria en la precicion del modelo luego de cambiar los parametros.
